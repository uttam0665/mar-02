```python
Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.
Ans:-
    
Matplotlib is a data visualization library for creating static, animated, and interactive visualizations in Python. 
It provides a wide variety of customizable and high-quality 2D and 3D plots, including line plots, scatter plots, bar plots, histograms, and more.

Matplotlib is used for a wide range of tasks in data science, including exploratory data analysis, data cleaning, modeling, and presentation.
It is widely used in scientific and engineering applications, as well as in finance, business, and social sciences.

The Pyplot module of Matplotlib provides a simple interface for creating various types of plots.
Here are five commonly used plots that can be plotted using the Pyplot module:

1.Line plot: A line plot is a plot that displays data as a series of points connected by straight lines. It is used to show trends over time or to compare data from different groups or categories.

2.Scatter plot: A scatter plot is a plot that displays data as a collection of points, where each point represents an observation in the data set. It is used to show the relationship between two variables.

3.Bar plot: A bar plot is a plot that displays data as a series of bars, where the height of each bar represents the value of a variable. It is used to compare the values of different categories or groups.

4.Histogram: A histogram is a plot that displays the distribution of a variable. It is used to show how frequently different values occur in a data set.

5.Pie chart: A pie chart is a plot that displays data as a circle divided into slices, where each slice represents a category or group and the size of each slice represents the proportion of observations in that category or group. It is used to show the relative sizes of different categories or groups.    
```

```python
"""Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot."""
#Ans:-

"""A scatter plot is a type of graph used to display the relationship between two variables. It is a visual representation of a set of data points, where each point represents the values of the two variables for a single observation.
The position of each point on the plot is determined by its x and y coordinates, with the x-axis typically representing the independent variable and the y-axis representing the dependent variable.
Scatter plots are useful for identifying patterns or trends in the data, as well as for detecting any outliers or unusual observations.
They can also be used to determine the strength and direction of the relationship between the two variables, which can be helpful in making predictions or in analyzing the effect of one variable on the other.
In a scatter plot, the points are often color-coded or labeled to represent additional information, such as the group or category to which each observation belongs.
This can make it easier to interpret the plot and to identify any patterns or relationships that may be present."""

import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.title("Scatter plot of x and y")
plt.xlabel("x axis")
plt.ylabel("x axis")
plt.show()
```

```python
"""Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])"""
# Ans:-
"""The subplot() function is used in Matplotlib, which is a popular data visualization library in Python, to create multiple plots or subplots within a single figure.
It allows you to divide the figure into a grid of rows and columns, and specify which subplot to draw on based on its position in the grid.
The subplot() function takes three arguments - the number of rows, the number of columns, and the index of the current subplot."""

import numpy as np
import matplotlib.pyplot as plt
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

fig.suptitle('Four Line Plots')
plt.show()
```

```python
"""Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)"""
#Ans:-
"""A box plot is a graphical representation of a data set that shows the distribution of the data. It is also known as a box-and-whisker plot. The box plot consists of a rectangular box with lines extending from it, called whiskers, that indicate variability outside the upper and lower quartiles of the data.
The box in a box plot represents the interquartile range (IQR), which is the range between the first and third quartiles of the data set. 
The line in the box represents the median of the data set. The whiskers extend from the box to the highest and lowest values that are still within 1.5 times the IQR from the box.
Any data points outside of the whiskers are considered outliers and are plotted as individual points.
Box plots are commonly used to visually summarize and compare the distribution of a variable across different groups or categories. 
They can be used to detect the presence of outliers, show the variability of the data, and provide a quick overview of the central tendency of the data. They are especially useful for comparing multiple data sets at once, as the box plots can be easily arranged side by side."""

import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]
fig, ax = plt.subplots()
ax.boxplot(data)
ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Value')
ax.set_title('Box Plot')
plt.show()

```
